```python
columns = ['cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']
for i in columns:
    sns.countplot(data1[i])
    plt.show()
    data1[i] = le.fit_transform(data1[i])
    sns.countplot(data1[i])
    plt.show()
```

```python
data1['class'] = le.fit_transform(data1['class'])
```

```python
data1.corr()
```

```python
sns.heatmap(data1.corr())
```

```python
y_data1 = data1['class']
X_data1 = data1.drop(['class'],axis = 1)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data1, y_data1, test_size = 0.30, random_state = 0)
```

```python
column = X_data1.columns
print(list(column))
```

```python
for c in columns:
    sns.barplot(x =X_data1[c] , y = data1['class'])
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor()
regr.fit(X_train,y_train)
print(regr.score(X_test,y_test))
```

```python
from sklearn import tree
clf = tree.DecisionTreeRegressor()
clf = clf.fit(X_train,y_train)
print(clf.score(X_test,y_test))
```

```python
import xgboost as xgb
xgb1 = xgb.XGBClassifier()
xgb1.fit(X_train,y_train)
print(xgb1.score(X_test,y_test))
```

```python
import lightgbm as lgb
clf = lgb.LGBMClassifier(random_state = 0)
clf.fit(X_train,y_train)
print(clf.score(X_test,y_test))
```
